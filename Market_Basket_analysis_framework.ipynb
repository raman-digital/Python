```python
SQL_Query = pd.read_sql_query(''' select ECOM_ORD_ID_PK, SIZE_DESCRIPTION,
 STYL_COLOR, ITM_ORD_QTY, 
 CATEGORY_NAME, DEPARTMENT_NAME, DEPARTMENT_DIVISION_NAME 
 from  temp_ecom_styl_col_size_qty_100 ''', conn)

```

```python
df = pd.DataFrame(SQL_Query, columns=['ECOM_ORD_ID_PK', 'SIZE_DESCRIPTION', 'STYL_COLOR', 'ITM_ORD_QTY', 'CATEGORY_NAME', 'DEPARTMENT_NAME', 'DEPARTMENT_DIVISION_NAME'])

print(df)
```

```python
basket1 = (df
          .groupby(['ECOM_ORD_ID_PK', 'SIZE_DESCRIPTION', 'STYL_COLOR'])['ITM_ORD_QTY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('ECOM_ORD_ID_PK'))
```

```python
basket2 = (df
          .groupby(['ECOM_ORD_ID_PK',  'STYL_COLOR'])['ITM_ORD_QTY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('ECOM_ORD_ID_PK'))
```

```python
print(basket2)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket2.applymap(encode_units)
#print(basket_sets)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemset, metric="lift", min_threshold=0.1)
rules.head()
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
rules.head()
```
